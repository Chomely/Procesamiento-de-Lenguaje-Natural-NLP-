## *Bucles:* Ciclo para iterar sobre un objeto. Comando que se repite varias veces a un objeto que tiene varios objetos. Cada vez que ve un objeto dentro de la lista se hace un acción.

```python
for _ in palabras:
    print(_)
```

```python
for aa in palabras:
    print(aa)
```

```python
for i,_ in enumerate(palabras): #enumerate Toma la lista y cada objeto lo enumera
    print(f"La palabra <<{_}>> está en el lugar #{i} de la lista")
```

```python
for numero in range(5,15):
    print(numero)
    print("jeje")
    
print("Jajaja") #Lenguaje -indentado
```

```python
for numero in range(5,15):
    if numero % 2 == 0: #Bucle con un condicional. Operador modulo
        print(f"{numero} es un número par")
    else:
        print(f"{numero} es un número impar")
```

Para poder leer tablas dentro del archivo

```python
tabla = documento.tables[0]
tabla
```

```python
data = [] #Creamos una lista vacia, y la vamos a iterar

for i, fila in enumerate(tabla.rows): #i cuenta los numeros yla fila cada objeto
    f = [celda.text for celda in fila.cells] #Recorro la fila, y luego cada celda de cada fila y lo agrego a la lista de comprensión 
    data.append(f)
    
df = pd.DataFrame(data)

df.columns = df.loc[0].values

df
```
