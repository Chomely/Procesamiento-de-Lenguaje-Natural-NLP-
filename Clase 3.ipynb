<div id="test" class="text" blah="valor">¡Hola, mundo!</div>

```python
ejemplo = bs.BeautifulSoup('<div id="test" class="text" blah="valor">¡Hola, mundo!</div>')
ejemplo
```

```python
# Cargar el documento
# Cargar el documento
ejemplo = bs.BeautifulSoup('''
                           <div id="test" class="text" blah="valor">¡Hola, mundo!</div>
                           <div id="test2" class="text" blah="valor">¡Hola, mundo 2!</div>
                           ''')
ejemplo

# Diferentes maneras de encontrar etiquetas
ejemplo.find_all("div") # Nombre de la etiqueta
ejemplo.find_all(class_="text") # Nombre y contenido del atributo
```

Requests

```python
import urllib.request
```

```python
link = "https://coinmarketcap.com/" #Tabla con nombres de criptomonedas 
```

```python
# Descargar la página web en Python 
request = urllib.request.Request(link, headers={'User-Agent': 'Mozilla/5.0'}) #Indico "soy" un navegador
webpage = urllib.request.urlopen(request) #Abro el link
source = webpage.read() #Obtener el código fuente
webpage.close()
  
```

```python
source
```

Procesos lapágina web con HTML

```python
soup = bs.BeautifulSoup(source, 'html.parser')
```
